```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('quikr_car.csv')
pd.DataFrame(df)
```

```python
df.shape
```

```python
df=df[df['year'].str.isnumeric()]
```

```python
df['year'] = df['year'].astype(int)
```

```python
df=df[df['Price']!='Ask For Price']
```

```python
# Convert 'Price' column to string and remove commas
df['Price'] = df['Price'].astype(str).str.replace(',', '')

# Convert 'Price' column to integer
df['Price'] = df['Price'].astype(int)

```

```python
df.sample(10)
```

```python
 df['kms_driven']= df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']= df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
df= df[~df['fuel_type'].isna()]
```

```python
df.shape
```

```python
df.info()
```

```python
df['name']= df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df=df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df =df[df['Price']<6e6].reset_index(drop=True)
```

```python
df.to_csv('cleaneddata.csv')
```

```python
#Model building
```

```python
X=df.drop(columns='Price')
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
columns_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(columns_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(columns_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(columns_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
